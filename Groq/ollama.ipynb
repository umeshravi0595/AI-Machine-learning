```python
import os
from dotenv import load_dotenv
load_dotenv()

## Langsmith Tracking
os.environ["LANGCHAIN_API_KEY"]=os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"]="true"
os.environ["LANGCHAIN_PROJECT"]=os.getenv("LANGCHAIN_PROJECT")
```

```python
from langchain_community.document_loaders import WebBaseLoader
```

```python
###This is data ingestion
loader=WebBaseLoader("https://docs.smith.langchain.com/observability/tutorials/observability")
loader
```

```python
doc=loader.load()
doc
```

```python
# splitting the text
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_split=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
doc_split=text_split.split_documents(doc)
doc_split
```
